## Calculate the mean
The arithmetic mean is the commonly used measure of central tendency. The Greek letter $\mu$ is used to represent the mean. To calculate the mean, we sum up all values $x_0+x_1+ ... x_n$ and divide it by the number of values $n$.

$$\mu=\frac{1}{n}\sum_ix_i$$


```python
# To calculate the mean ourselves
x, n = 0.0, 0

for number_of_steps in data.steps:
    x += number_of_steps
    n += 1

mean = x / n
mean
```

```python
# We can also let pandas use NumPy's mean function to do the job
data.steps.mean()
```

```python
# Or we can call NumPy's mean function ourselves
np.mean(data.steps)
```

```python
# Use pandas to get the mean for all columns at ones
data.mean()
```

## Calculate the median
The median is often a better measure of central tendency when we have extreme outliers. The median is the value in the middle after we sort the data. If the number of observations $n$ is even, we have to take the mean of the two middle values. We assume a zero based index.


$$n\ is\ odd:\ \ x_{median}=x_{\frac{n-1}{2}}$$

$$n\ is\ even:\ \ x_{median}=\frac{x_\frac{n-2}{2}+x_\frac{n}{2}}{2}$$


```python
median, n = 0.0, 0

# Count the number of observations
for number_of_steps in data.steps:
    n += 1
    
# order the data
ordered_data = data.steps.order()

if n % 2 == 0:
    # n is even
    m1 = ordered_data.ix[(n - 2) / 2]
    m2 = ordered_data.ix[(n / 2)]
    median = (m1 + m2) / 2.0
else:
    # n is odd
    median = ordered_data.ix[(n - 2) / 2.0]

median
```

```python
# Again, we can let pandas use NumPy's median function to do the job
data.steps.median()
```

```python
# Or we can call NumPy's median function ourselves
np.median(data.steps)
```

```python
# Use pandas to get the median for all columns at ones
data.median()
```
